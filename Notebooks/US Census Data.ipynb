```python
import pandas as pd
import glob
import os
```

```python
# Import xlsx file
xlsx_file = '../Resources/cbsa-met-est2021-pop.xlsx'

# Skip rows to reduce the merged cells in the file
census_2021_df = pd.read_excel(xlsx_file, skiprows = range(1,4))
#census_2021_df = pd.read_excel(xlsx_file)

census_2021_df.head()
```

```python
# rename columns
census_2021_df=census_2021_df.rename(columns = {
    'table with row headers in column A and column headers in rows 3 through 4 (leading dots indicate sub-parts)':'Geographic Area',
    'Unnamed: 1':'April 2020',
    'Unnamed: 2':'2020',
    'Unnamed: 3':'2021'
})
census_2021_df
```

```python
census_2021_df = census_2021_df.dropna()
census_2021_df
```

```python
us_cities_2021 = ['Atlanta',
                    'Austin',
                    'Baltimore',
                    'Boston',
                    'Buffalo',
                    'Charlotte',
                    'Chicago',
                    'Cincinnati',
                    'Cleveland',
                    'Columbus',
                    'Dallas',
                    'Denver',
                    'Detroit',
                    'Green Bay',
                    'Houston',
                    'Indianapolis',
                    'Jacksonville',
                    'Kansas City',
                    'Las Vegas',
                    'Los Angeles',
                    'Memphis',
                    'Miami',
                    'Minneapolis',
                    'Milwaukee',
                    'Nashville',
                    'New Orleans',
                    'New York',
                    'Oakland',
                    'Oklahoma City',
                    'Orlando',
                    'Philadelphia',
                    'Phoenix',
                    'Pittsburgh',
                    'Portland',
                    'Raleigh',
                    'Sacramento',
                    'Salt Lake',
                    'San Antonio',
                    'San Diego',
                    'San Francisco',
                    'San Jose',
                    'St. Louis',
                    'Seattle',
                    'Tampa',
                    'Washington']

searchcity = '|'.join(us_cities_2021)
pro_us_cities_2021 = census_2021_df[census_2021_df['Geographic Area'].str.contains(searchcity)]
pro_us_cities_2021

# pro_us_cities_2021 = census_2021_df.loc[census_2021_df['Geographic Area'].isin(us_cities_2021)]
# pro_us_cities_2021
```

```python
new_df = pd.DataFrame(columns = ['Geographic Area', 'April 2020', '2020', '2021'])
temp_df = pd.DataFrame()
for city in us_cities_2021:
    temp_df = pro_us_cities_2021.loc[pro_us_cities_2021['Geographic Area'].str.contains(city)]
    temp_df = temp_df.sort_values('2021',ascending=False)
    temp_df['City'] = city
    new_df=new_df.append(temp_df.head(1))
    
final_census_21=new_df.set_index('City')
final_census_21
```

```python
final_census_21.to_csv('us_census_data_2021.csv', index=True)
```
