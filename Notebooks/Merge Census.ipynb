```python
import pandas as pd
import glob
import os
```

```python
csv_file = '../Resources/CanadaSportsCities.csv'
census_df = pd.read_csv(csv_file)
census_df.head()
```

```python
# rename columns
census_df=census_df.rename(columns = {
    'GEO':'City'
})
census_df
```

```python
clean_canada=census_df[['City','2016 Population', '2021 Population']]
clean_canada['Country Code']='CAN'
clean_canada.head()
```

```python
usa_csv_file = '../Resources/us_census_data_merged.csv'
usa_census_df = pd.read_csv(usa_csv_file)
usa_census_df.head()
```

```python
# rename columns
usa_census_df=usa_census_df.rename(columns = {
    '2016':'2016 Population',
    '2021':'2021 Population',
})
usa_census_df
```

```python
clean_usa=usa_census_df[['City','2016 Population', '2021 Population']]
clean_usa['Country Code']='USA'
clean_usa.head()
```

```python
final_merge_df=pd.concat([clean_usa, clean_canada])
final_merge_df.head()
```

```python
final_merge_df = final_merge_df.astype({
    '2016 Population':int,
    '2021 Population':int
})
final_merge_df.head()
```

```python
final_merge_df.to_csv('../Resources/final_country_census.csv', index=False)
```
