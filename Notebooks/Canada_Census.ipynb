```python
import pandas as pd
```

```python
csv_file = '../Resources/98100003.csv'
census_df = pd.read_csv(csv_file)
census_df.head()
```

```python
cities_list = ['Calgary','Edmonton','Montreal','Ottawa','Toronto','Vancouver','Winnipeg']
filtered_cities = census_df.loc[census_df['GEO'].isin(cities_list)]
filtered_cities
```

```python
group_cities = filtered_cities.groupby(['GEO'])
sports_cities = group_cities.max()
sports_cities
```

```python
reduced_sports_cities = sports_cities[['REF_DATE','Population and dwelling counts (13): Population, 2021 [1]',
                                      'Population and dwelling counts (13): Population, 2016 [2]',
                                     'Population and dwelling counts (13): Population percentage change, 2016 to 2021 [3]',
                                      'Population and dwelling counts (13): Population density per square kilometre, 2021 [11]']]

reduced_sports_cities
```

```python
reduced_sports_cities = reduced_sports_cities.rename(columns={
    'REF_DATE':'Census Year',
    'Population and dwelling counts (13): Population, 2021 [1]':'2021 Population',
    'Population and dwelling counts (13): Population, 2016 [2]':'2016 Population',
    'Population and dwelling counts (13): Population percentage change, 2016 to 2021 [3]':'% change from 2016 to 2021',
    'Population and dwelling counts (13): Population density per square kilometre, 2021 [11]':'Population Density per sq km'

})
reduced_sports_cities
```

```python
# convert from sq km to sq mi
reduced_sports_cities['Population Density per sq mi'] = round(reduced_sports_cities['Population Density per sq km'] * (0.621371 **2),1)
reduced_sports_cities
```

```python
reduced_sports_cities.to_csv('../Resources/CanadaSportsCities.csv')
```
